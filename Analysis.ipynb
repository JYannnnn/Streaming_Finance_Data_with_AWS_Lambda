# Streaming Finance Data with AWS Lambda - Data Visualizations

## Import Data:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("~/Desktop/result.csv")
df
```

## Graph 1: Average Daily Highest Stock Price by Company

```python
sns.catplot(x="company",y="high_stock_price",data=df,kind="bar",palette="ch:.25",ci=None)
plt.title('Average Daily Highest Stock Price by Company')
plt.show()
```

## Graph 2: Hourly Highest Stock Price by Company

```python
g = sns.relplot(y="high_stock_price",x="hour",data=df,kind="line",hue="company")
g.fig.suptitle("Hourly Highest Stock Price by Company",y=1.05)
plt.show()
```

## Graph 3: Highest Stock Price Comparison Between Hours 

```python
g = sns.catplot(x="hour",y="high_stock_price",data=df,kind="box",palette="GnBu")
g.fig.suptitle("Highest Stock Price Comparison Between Hours",y=1.05)
plt.show()
```
